```python
#funciones de limpieza
def format_columns(df):
    df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
    df = df.rename(columns = {'st': 'state'})
    return df

def print_unique(df): #valores unicos de una columna
    for column in ['state', 'gender', 'policy_type', 'vehicle_class']:
        print(f'{column}: {df[column].unique()}')

def replace(df):
    return df.replace({
        'state': {'AZ': 'Arizona', 'WA': 'Washington', 'Cali': 'California', 'CA': 'California'},
        'gender': {'female': 'F', 'Male': 'M', 'Femal': 'F'},
        'vehicle_class': {'Luxury SUV': 'Luxury', 'Luxury Car': 'Luxury', 'SUV': 'Luxury'},
        'education': {'Bachelors': 'Bachelor'}
    })
```

```python
data1= format_columns(data1)
data2 = format_columns(data2)
data3 = format_columns(data3)

print_unique(data1)
data1 = replace(data1)

print('\n')
print_unique(data2)
data2 = replace(data2)

print('\n')
print_unique(data3)
data3 = replace(data3)
```

```python


def data_type_formatting(df):
    # Manejar posibles valores nulos y convertir customer_lifetime_value a float
    df['customer_lifetime_value'] = pd.to_numeric(df['customer_lifetime_value'].str.rstrip('%'), errors='coerce')
    
    # Extraer el segundo valor de la fracción en number_of_open_complaints y convertirlo a float
    df['number_of_open_complaints'] = df['number_of_open_complaints'].str.split('/').str[1]
    df['number_of_open_complaints'] = pd.to_numeric(df['number_of_open_complaints'], errors='coerce')

    return df
```

```python
print(data1.customer.isnull().sum())
print(data2.customer_lifetime_value.isnull().sum())
print(data3.customer_lifetime_value.isnull().sum())
```

```python
data1.dropna(subset = ['customer','customer_lifetime_value'], inplace = True) # elimina valores nulos, y corrige el d.f origanal
```

```python
data1.gender.fillna(data1.gender.mode()[0], inplace = True) #rellenar valores nulos
```

```python
data2.dropna(subset = ['customer_lifetime_value'], inplace = True)
```

```python
df_final = pd.concat([data1, data2, data3], ignore_index=True)  

```

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv')
df1.head()
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
pivot_df1 = df1.pivot_table(index = 'sales_channel', values = ['customer_lifetime_value'], aggfunc='sum')
pivot_df1 = pivot_df1.customer_lifetime_value.apply(lambda y: "{:.2e}".format(y))
pivot_df1 
# Agent es la forma que trae el total de ingresos mas alto, seguido de Branch, Call Center, y Web. 
# Agent trae el doble de ingresos en comparación the Call Center branch. :)
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python

pivot1_df1 = df1.pivot_table(index = 'gender', columns = 'education', values = ['customer_lifetime_value'], aggfunc = 'mean')
pivot1_df1 = pivot1_df1.customer_lifetime_value.apply(lambda y: round(y, 2))
pivot1_df1
# Gente con Master y High School or Below education representan el mas alto nivel de ingresos. 
# Gente con Bachelor, College and Doctor education muestran un nivel de ingreso similar.
# Females con  High School or Below education superan en sus ingresos por 500$ a los  males.
# Males  con College education tienen en total un ingreso superior de 300$ comparado con females.
```
